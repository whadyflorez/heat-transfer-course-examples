# Transferencia de calor
## Convección forzada externa sobre un cilindro en flujo cruzado

### Aplicación real en ingeniería mecánica

En una planta de procesos, una **tubería horizontal de acero** transporta un fluido caliente y atraviesa una zona donde se ha instalado un sistema de ventilación forzada para disminuir la temperatura superficial exterior y proteger al personal de contacto accidental.

La tubería puede modelarse como un **cilindro largo en flujo cruzado**. Aire ambiente a **20 °C** y **1 atm** es impulsado perpendicularmente al eje de la tubería mediante un ventilador, con una velocidad uniforme de **8 m/s**.

La tubería tiene un **diámetro exterior** de $D=0.060\ \mathrm{m}$ y una **longitud expuesta al aire** de $L=4.0\ \mathrm{m}$. Durante operación estacionaria, la temperatura superficial exterior de la tubería se mantiene aproximadamente constante en $T_s=110\ ^\circ\mathrm{C}$.

Se desea estimar la tasa de pérdida de calor por convección desde la superficie exterior de la tubería hacia el aire.


## Objetivos

1. Calcular la **temperatura de película**.
2. Obtener con **CoolProp** las propiedades del aire a la temperatura de película.
3. Calcular $\nu$, $Pr$ y $Re_D$.
4. Usar una correlación apropiada para **flujo cruzado sobre cilindros**.
5. Determinar el número de Nusselt promedio, el coeficiente convectivo $h$, el área lateral y la tasa de pérdida de calor.
6. Interpretar los resultados desde el punto de vista de la ingeniería mecánica.


## Datos del problema

$$
T_\infty = 20\ ^\circ\mathrm{C}
$$

$$
T_s = 110\ ^\circ\mathrm{C}
$$

$$
P = 101325\ \mathrm{Pa}
$$

$$
U_\infty = 8\ \mathrm{m/s}
$$

$$
D = 0.060\ \mathrm{m}
$$

$$
L = 4.0\ \mathrm{m}
$$


## Fundamento teórico

La **temperatura de película** se define como:

$$
T_f = \frac{T_s + T_\infty}{2}
$$

El número de Reynolds para un cilindro en flujo cruzado se calcula con el **diámetro** como longitud característica:

$$
Re_D = \frac{U_\infty D}{\nu}
$$

donde:

$$
\nu = \frac{\mu}{\rho}
$$

y el número de Prandtl se calcula como:

$$
Pr = \frac{\mu c_p}{k}
$$

Para un cilindro largo en flujo cruzado, una correlación ampliamente usada es la de **Churchill y Bernstein**:

$$
Nu_D = 0.3 + \frac{0.62\,Re_D^{1/2}Pr^{1/3}}{\left[1+\left(0.4/Pr\right)^{2/3}\right]^{1/4}}\left[1+\left(\frac{Re_D}{282000}\right)^{5/8}\right]^{4/5}
$$

Luego, el coeficiente convectivo promedio se obtiene de:

$$
h = \frac{Nu_D\,k}{D}
$$

El área lateral externa de la tubería es:

$$
A_s = \pi D L
$$

y la tasa de transferencia de calor por convección es:

$$
\dot{Q} = h A_s (T_s - T_\infty)
$$

Finalmente, el flujo de calor promedio es:

$$
q'' = \frac{\dot{Q}}{A_s}
$$

````python
from CoolProp.CoolProp import PropsSI
import math
````

## 1. Datos del problema y conversión de unidades


````python
# Datos
T_inf_C = 20.0       # °C
T_s_C   = 110.0      # °C
P       = 101325.0   # Pa
U       = 8.0        # m/s
D       = 0.060      # m
L       = 4.0        # m

# Conversión a Kelvin
T_inf = T_inf_C + 273.15
T_s   = T_s_C + 273.15

print(f"T_inf = {T_inf:.2f} K")
print(f"T_s   = {T_s:.2f} K")
````

## 2. Temperatura de película


````python
T_f = 0.5 * (T_inf + T_s)
print(f"Temperatura de película = {T_f:.2f} K")
print(f"Temperatura de película = {T_f - 273.15:.2f} °C")
````

## 3. Propiedades del aire con CoolProp

Se evalúan las propiedades del aire a la temperatura de película y a 1 atm:

- $\rho$: densidad
- $\mu$: viscosidad dinámica
- $k$: conductividad térmica
- $c_p$: calor específico a presión constante


````python
rho = PropsSI('D', 'T', T_f, 'P', P, 'Air')   # kg/m^3
mu  = PropsSI('V', 'T', T_f, 'P', P, 'Air')   # Pa.s
k   = PropsSI('L', 'T', T_f, 'P', P, 'Air')   # W/m.K
cp  = PropsSI('C', 'T', T_f, 'P', P, 'Air')   # J/kg.K

print('Propiedades del aire a la temperatura de película:')
print(f"rho = {rho:.6f} kg/m^3")
print(f"mu  = {mu:.6e} Pa·s")
print(f"k   = {k:.6f} W/m·K")
print(f"cp  = {cp:.2f} J/kg·K")
````

## 4. Propiedades derivadas

Calculamos la viscosidad cinemática y el número de Prandtl:

$$
\nu = \frac{\mu}{\rho}
$$

$$
Pr = \frac{\mu c_p}{k}
$$


````python
nu = mu / rho
Pr = mu * cp / k

print(f"nu = {nu:.6e} m^2/s")
print(f"Pr = {Pr:.6f}")
````

## 5. Número de Reynolds

Para flujo cruzado sobre un cilindro:

$$
Re_D = \frac{U_\infty D}{\nu}
$$


````python
Re_D = U * D / nu
print(f"Re_D = {Re_D:.6e}")
````

## 6. Número de Nusselt promedio

Usamos la correlación de Churchill y Bernstein:

$$
Nu_D = 0.3 + \frac{0.62\,Re_D^{1/2}Pr^{1/3}}{\left[1+\left(0.4/Pr\right)^{2/3}\right]^{1/4}}\left[1+\left(\frac{Re_D}{282000}\right)^{5/8}\right]^{4/5}
$$


````python
Nu_D = 0.3 + (
    (0.62 * Re_D**0.5 * Pr**(1/3)) /
    ((1 + (0.4 / Pr)**(2/3))**0.25)
) * (1 + (Re_D / 282000)**(5/8))**(4/5)

print(f"Nu_D = {Nu_D:.6f}")
````

## 7. Coeficiente convectivo promedio

$$
h = \frac{Nu_D\,k}{D}
$$


````python
h = Nu_D * k / D
print(f"h = {h:.6f} W/m^2·K")
````

## 8. Área lateral externa de la tubería

$$
A_s = \pi D L
$$


````python
A = math.pi * D * L
print(f"A_s = {A:.6f} m^2")
````

## 9. Tasa de transferencia de calor y flujo de calor promedio

$$
\dot{Q} = h A_s (T_s - T_\infty)
$$

$$
q'' = \frac{\dot{Q}}{A_s}
$$


````python
Q = h * A * (T_s - T_inf)
q_pp = Q / A

print(f"Q = {Q:.6f} W")
print(f"q'' = {q_pp:.6f} W/m^2")
````

## 10. Resumen organizado de resultados


````python
print('=== RESUMEN DE RESULTADOS ===')
print(f"Temperatura de película      : {T_f - 273.15:.2f} °C")
print(f"Densidad rho                 : {rho:.6f} kg/m^3")
print(f"Viscosidad dinámica mu       : {mu:.6e} Pa·s")
print(f"Conductividad térmica k      : {k:.6f} W/m·K")
print(f"Calor específico cp          : {cp:.2f} J/kg·K")
print(f"Viscosidad cinemática nu     : {nu:.6e} m^2/s")
print(f"Número de Prandtl Pr         : {Pr:.6f}")
print(f"Número de Reynolds Re_D      : {Re_D:.6e}")
print(f"Número de Nusselt Nu_D       : {Nu_D:.6f}")
print(f"Coeficiente convectivo h     : {h:.6f} W/m^2·K")
print(f"Área lateral A_s             : {A:.6f} m^2")
print(f"Tasa de calor Q              : {Q:.6f} W")
print(f"Flujo de calor promedio q''  : {q_pp:.6f} W/m^2")
````
