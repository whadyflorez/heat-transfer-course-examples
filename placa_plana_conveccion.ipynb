# Transferencia de Calor
## Convección Forzada Externa sobre Placa Plana

Este notebook desarrolla un ejercicio de convección forzada externa usando correlaciones clásicas y propiedades obtenidas con CoolProp.

---

## Problema

Aire a 25 °C y 1 atm fluye sobre una placa plana de longitud 0.80 m y ancho 0.30 m con velocidad de 6 m/s.

La placa está a temperatura constante de 80 °C.

Se desea calcular:

1. Temperatura de película
2. Propiedades del aire con CoolProp
3. Reynolds y Prandtl
4. Nusselt promedio
5. Coeficiente convectivo
6. Calor transferido

---

## 1. Datos del problema

````python
# Datos
T_inf_C = 25.0
T_s_C = 80.0
P = 101325.0
U = 6.0
L = 0.80
b = 0.30

# Conversión a Kelvin
T_inf = T_inf_C + 273.15
T_s = T_s_C + 273.15
````

## 2. Temperatura de película

````python
T_f = 0.5 * (T_inf + T_s)
print(f"Temperatura de película: {T_f - 273.15:.2f} °C")
````

## 3. Propiedades con CoolProp

Instalar si es necesario:
```
pip install CoolProp
```

````python
from CoolProp.CoolProp import PropsSI

rho = PropsSI('D', 'T', T_f, 'P', P, 'Air')
mu  = PropsSI('V', 'T', T_f, 'P', P, 'Air')
k   = PropsSI('L', 'T', T_f, 'P', P, 'Air')
cp  = PropsSI('C', 'T', T_f, 'P', P, 'Air')

print("Propiedades del aire:")
print(f"rho = {rho:.4f} kg/m^3")
print(f"mu  = {mu:.6e} Pa·s")
print(f"k   = {k:.5f} W/m·K")
print(f"cp  = {cp:.2f} J/kg·K")
````

## 4. Cálculo de propiedades derivadas

````python
nu = mu / rho
Pr = mu * cp / k

print(f"nu = {nu:.6e} m^2/s")
print(f"Pr = {Pr:.4f}")
````

## 5. Número de Reynolds

````python
Re_L = U * L / nu
print(f"Re_L = {Re_L:.4e}")
````

## 6. Número de Nusselt promedio (flujo laminar)

$$
Nu_L = 0.664 Re_L^{1/2} Pr^{1/3}
$$

````python
Nu_L = 0.664 * (Re_L ** 0.5) * (Pr ** (1/3))
print(f"Nu_L = {Nu_L:.2f}")
````

## 7. Coeficiente convectivo

````python
h = Nu_L * k / L
print(f"h = {h:.2f} W/m^2·K")
````

## 8. Calor transferido

````python
A = L * b
Q = h * A * (T_s - T_inf)
q_pp = Q / A

print(f"Área = {A:.3f} m^2")
print(f"Q = {Q:.2f} W")
print(f"q'' = {q_pp:.2f} W/m^2")
````

## 9. Análisis del régimen de flujo

- Si Re_L < 5e5 → flujo laminar
- Si Re_L > 5e5 → transición o turbulento


````python
if Re_L < 5e5:
    print("Flujo laminar en toda la placa")
else:
    print("Hay transición o flujo turbulento")
````

## 10. Extensión: velocidad mayor

Repetir el cálculo con U = 18 m/s

````python
U2 = 18.0

Re_L2 = U2 * L / nu

if Re_L2 < 5e5:
    Nu_L2 = 0.664 * (Re_L2 ** 0.5) * (Pr ** (1/3))
else:
    Nu_L2 = (0.037 * (Re_L2 ** 0.8) - 871) * (Pr ** (1/3))

h2 = Nu_L2 * k / L
Q2 = h2 * A * (T_s - T_inf)

print(f"Re_L = {Re_L2:.4e}")
print(f"Nu_L = {Nu_L2:.2f}")
print(f"h = {h2:.2f} W/m^2·K")
print(f"Q = {Q2:.2f} W")
````

## Conclusiones

- Uso de temperatura de película es fundamental
- CoolProp permite propiedades precisas
- La selección de correlación depende del régimen
- El aumento de velocidad incrementa significativamente la transferencia de calor
