Ejemplo propuesto: flujo interno con calentamiento por condensación externa
---------------------------------------------------------------------------

Se desea calentar agua que circula por el interior de una tubería horizontal de diámetro  $D$  y longitud  $L$ . En el exterior de la tubería hay vapor saturado que condensa sobre la superficie externa. Se pide determinar:

1.  El coeficiente de convección interno  $h_i$ .
2.  El coeficiente promedio de condensación externo  $h_o$ .
3.  La temperatura de pared  $T_s$ , asumida constante pero desconocida.
4.  La temperatura de salida del fluido interno  $T_{out}$ .

### Supuestos

*   Régimen permanente.
*   Flujo unidimensional en el interior.
*   Propiedades constantes.
*   Espesor de pared despreciable, por lo que la temperatura de la pared es la misma vista desde adentro y desde afuera.
*   La pared se mantiene a temperatura uniforme  $T_s$ , pero esa temperatura no se conoce a priori.
*   El vapor externo está a temperatura de saturación constante  $T_{sat}$ .
*   La condensación exterior es laminar en película sobre un tubo horizontal.

* * *

Datos del problema
------------------

Agua fluye por el interior de una tubería de acero delgado con los siguientes datos:

$$
D = 0.020\ \text{m}
$$
 
$$
L = 5.0\ \text{m}
$$
 
$$
\dot m = 0.15\ \text{kg/s}
$$
 
$$
T_{in} = 20^\circ C
$$

En el exterior condensa vapor saturado a:

$$
T_{sat} = 100^\circ C
$$

Para el agua interna, evaluar propiedades a una temperatura media aproximada de  $40^\circ C$ :

$$
c_p = 4180\ \text{J/(kg·K)}, \qquad k = 0.63\ \text{W/(m·K)}
$$
 
$$
\mu = 7.0\times 10^{-4}\ \text{Pa·s}, \qquad Pr = 4.7
$$

Para el condensado externo de agua cerca de  $100^\circ C$ , usar:

$$
\rho_l = 958\ \text{kg/m}^3
$$
 
$$
\rho_v = 0.60\ \text{kg/m}^3
$$
 
$$
\mu_l = 2.82\times 10^{-4}\ \text{Pa·s}
$$
 
$$
k_l = 0.679\ \text{W/(m·K)}
$$
 
$$
h_{fg} = 2.257\times 10^6\ \text{J/kg}
$$
 
$$
g = 9.81\ \text{m/s}^2
$$

* * *

1) Ecuación de energía para el fluido interno
---------------------------------------------

Tomemos un elemento diferencial de longitud  $dx$ . El calor que entra al fluido desde la pared es:

$$
\delta \dot Q = h_i P \left(T_s - T\right)\, dx
$$

donde  $P=\pi D$  es el perímetro interno.

Ese calor produce el incremento de entalpía del fluido:

$$
\dot m c_p\, dT = h_i P (T_s-T)\, dx
$$

Por tanto, la ecuación diferencial correcta es:

$$
\dot m c_p \frac{dT}{dx} = h_i P (T_s-T)
$$

o también:

$$
\dot m c_p \frac{dT}{dx} + h_i P (T-T_s)=0
$$

Esa es la forma correcta.  
La que escribiste originalmente

$$
\dot m c_p \frac{dT}{dx}+hP(T_s-T)=0
$$

tiene el signo inconsistente si el fluido se está calentando.

* * *

2) Solución de la ecuación diferencial
--------------------------------------

Separando variables:

$$
\frac{dT}{T_s-T}=\frac{h_iP}{\dot m c_p}\,dx
$$

Integrando desde  $x=0$ , donde  $T=T_{in}$ , hasta una posición genérica  $x$ :

$$
T(x)=T_s-\left(T_s-T_{in}\right)\exp\!\left(-\frac{h_iPx}{\dot m c_p}\right)
$$

En particular, a la salida  $x=L$ :

$$
T_{out}=T_s-\left(T_s-T_{in}\right)\exp\!\left(-\frac{h_iPL}{\dot m c_p}\right)
$$

Esta es la primera ecuación clave, y efectivamente deja  $T_{out}$  en función de  $T_s$ .

* * *

3) Coeficiente de convección interno  $h_i$ 
--------------------------------------------

Primero calculamos el número de Reynolds para el flujo interno:

$$
Re=\frac{4\dot m}{\pi D \mu}
$$

Sustituyendo:

$$
Re=\frac{4(0.15)}{\pi(0.020)(7.0\times 10^{-4})}\approx 1.36\times 10^4
$$

Como  $Re>10^4$ , podemos usar Dittus-Boelter para flujo turbulento interno:

$$
Nu = 0.023\,Re^{0.8}Pr^{0.4}
$$

Entonces:

$$
Nu = 0.023(1.36\times 10^4)^{0.8}(4.7)^{0.4}\approx 86.8
$$

Y por definición:

$$
h_i=\frac{Nu\,k}{D}
$$
 
$$
h_i=\frac{(86.8)(0.63)}{0.020}\approx 2.77\times 10^3\ \text{W/(m}^2\text{·K)}
$$

Por tanto:

$$
\boxed{h_i \approx 2.77\times 10^3\ \text{W/(m}^2\text{·K)}}
$$

* * *

4) Expresión de  $T_{out}$ 
---------------------------

Como

$$
P=\pi D = \pi(0.020)=0.06283\ \text{m}
$$

entonces

$$
\frac{h_iPL}{\dot m c_p} = \frac{(2766)(0.06283)(5)}{(0.15)(4180)} \approx 1.386
$$

Luego:

$$
T_{out}=T_s-(T_s-20)\,e^{-1.386}
$$

Como

$$
e^{-1.386}\approx 0.250
$$

queda:

$$
T_{out}=T_s-0.25(T_s-20)
$$
 
$$
T_{out}=0.75\,T_s+5
$$

Esta expresión es muy útil para la clase porque deja ver inmediatamente que la salida depende linealmente de la temperatura de pared si  $h_i$  ya fue estimado.

* * *

5) Calor total recibido por el fluido interno
---------------------------------------------

Por termodinámica:

$$
\dot Q_{fluido}=\dot m c_p (T_{out}-T_{in})
$$

Sustituyendo  $T_{in}=20^\circ C$ :

$$
\dot Q_{fluido}=(0.15)(4180)(T_{out}-20)
$$
 
$$
\dot Q_{fluido}=627(T_{out}-20)
$$

Pero como  $T_{out}=0.75T_s+5$ ,

$$
\dot Q_{fluido}=627(0.75T_s+5-20)
$$
 
$$
\dot Q_{fluido}=627(0.75T_s-15)
$$
 
$$
\boxed{\dot Q_{fluido}=470.25(T_s-20)}
$$

* * *

6) Coeficiente de condensación externo
--------------------------------------

Para condensación laminar en película sobre un tubo horizontal, la correlación de Nusselt es:

$$
\bar h_o = 0.725 \left[ \frac{\rho_l(\rho_l-\rho_v)g h_{fg} k_l^3} {\mu_l D (T_{sat}-T_s)} \right]^{1/4}
$$

Observa que  $h_o$  depende de  $T_s$ . Eso hace que el problema sea acoplado y no lineal.

Sustituyendo todas las propiedades excepto  $(T_{sat}-T_s)$ :

$$
\bar h_o = 0.725 \left[ \frac{(958)(958-0.60)(9.81)(2.257\times10^6)(0.679)^3} {(2.82\times10^{-4})(0.020)(100-T_s)} \right]^{1/4}
$$

Agrupando constantes:

$$
\bar h_o \approx 23623\,(100-T_s)^{-1/4}
$$

con  $T_s$  en  $^\circ C$  o K, porque solo aparece una diferencia de temperaturas.

* * *

7) Calor liberado por la condensación
-------------------------------------

El calor transferido desde el vapor al tubo es:

$$
\dot Q_{cond}=\bar h_o A_o (T_{sat}-T_s)
$$

Como el espesor es despreciable, podemos tomar:

$$
A_o \approx \pi D L
$$
 
$$
A_o = \pi(0.020)(5)=0.3142\ \text{m}^2
$$

Entonces:

$$
\dot Q_{cond} = 23623(100-T_s)^{-1/4}(0.3142)(100-T_s)
$$
 
$$
\dot Q_{cond} = 7419(100-T_s)^{3/4}
$$

Por tanto:

$$
\boxed{\dot Q_{cond}=7419(100-T_s)^{3/4}}
$$

* * *

8) Segunda ecuación: balance global de energía
----------------------------------------------

La segunda ecuación necesaria es justamente la igualdad entre el calor que cede el vapor al condensar y el calor que gana el fluido interno:

$$
\dot Q_{cond}=\dot Q_{fluido}
$$

Entonces:

$$
7419(100-T_s)^{3/4}=470.25(T_s-20)
$$

Esta ecuación no lineal permite obtener  $T_s$ .

* * *

9) Solución para  $T_s$ 
------------------------

Resolviendo numéricamente:

$$
\boxed{T_s \approx 92.4^\circ C}
$$

* * *

10) Temperatura de salida del fluido
------------------------------------

Usando

$$
T_{out}=0.75T_s+5
$$

se obtiene:

$$
T_{out}=0.75(92.4)+5
$$
 
$$
\boxed{T_{out}\approx 74.3^\circ C}
$$

* * *

11) Verificación del balance de energía
---------------------------------------

### Calor ganado por el fluido

$$
\dot Q_{fluido} = \dot m c_p (T_{out}-T_{in})
$$
 
$$
\dot Q_{fluido} = (0.15)(4180)(74.3-20) \approx 3.40\times 10^4\ \text{W}
$$
 
$$
\boxed{\dot Q_{fluido}\approx 34.0\ \text{kW}}
$$

### Calor cedido por condensación

$$
\dot Q_{cond}=7419(100-92.4)^{3/4}
$$
 
$$
\dot Q_{cond}\approx 34.0\ \text{kW}
$$

El balance cierra correctamente.

* * *

Resultado final del ejemplo
---------------------------

$$
\boxed{h_i \approx 2.77\times 10^3\ \text{W/(m}^2\text{·K)}}
$$
 
$$
\boxed{T_s \approx 92.4^\circ C}
$$
 
$$
\boxed{T_{out} \approx 74.3^\circ C}
$$
 
$$
\boxed{\dot Q \approx 34.0\ \text{kW}}
$$

* * *

Qué enseña este ejemplo
-----------------------

Este ejemplo es muy bueno para clase porque integra varias ideas a la vez:

*   uso de correlaciones de convección interna;
*   uso de correlaciones de condensación;
*   planteamiento correcto de la ecuación diferencial de energía;
*   solución analítica de la ODE;
*   acoplamiento no lineal entre transferencia interna y externa;
*   cierre del problema con un balance global de energía.

Además, muestra algo importante: aunque la pared tenga temperatura constante, esa temperatura no tiene por qué ser conocida. Se determina como parte de la solución.

* * *

Forma general del procedimiento para tus estudiantes
----------------------------------------------------

1.  Calcular  $Re$ ,  $Pr$ ,  $Nu$  y  $h_i$ .
2.  Resolver la ecuación diferencial interna y obtener  $T_{out}(T_s)$ .
3.  Escribir  $\dot Q_{fluido}=\dot m c_p(T_{out}-T_{in})$ .
4.  Escribir la correlación de condensación para  $h_o(T_s)$ .
5.  Escribir  $\dot Q_{cond}=h_o(T_s)A_o(T_{sat}-T_s)$ .
6.  Igualar  $\dot Q_{cond}=\dot Q_{fluido}$  y resolver para  $T_s$ .
7.  Sustituir en la expresión de  $T_{out}$ .

* * *

Observación didáctica importante
--------------------------------

Este modelo supone que la resistencia dominante es convectiva y que la resistencia de la pared es despreciable. Si más adelante quieres sofisticarlo, se puede extender a:

$$
\frac{1}{U A}= \frac{1}{h_i A_i} + R_{pared} + \frac{1}{h_o A_o}
$$

y entonces ya no se asume una sola temperatura de pared ni espesor despreciable.

```python
import numpy as np
from scipy.optimize import fsolve

# ============================================================
# DATOS DEL PROBLEMA
# ============================================================

# Geometría de la tubería
D = 0.020       # m
L = 5.0         # m
P = np.pi * D   # perímetro interno/externo aproximado
A = np.pi * D * L  # área lateral de la tubería

# Fluido interno: agua
m_dot = 0.15        # kg/s
T_in = 20.0         # °C
cp = 4180.0         # J/(kg·K)
k_i = 0.63          # W/(m·K)
mu_i = 7.0e-4       # Pa·s
Pr = 4.7            # adimensional

# Vapor externo en condensación
T_sat = 100.0       # °C

# Propiedades del condensado (agua cerca de 100 °C)
rho_l = 958.0       # kg/m^3
rho_v = 0.60        # kg/m^3
mu_l = 2.82e-4      # Pa·s
k_l = 0.679         # W/(m·K)
h_fg = 2.257e6      # J/kg
g = 9.81            # m/s^2


# ============================================================
# 1. COEFICIENTE DE CONVECCIÓN INTERNO
# ============================================================

# Reynolds para flujo interno en tubería circular
Re = 4.0 * m_dot / (np.pi * D * mu_i)

# Correlación de Dittus-Boelter
Nu = 0.023 * Re**0.8 * Pr**0.4

# Coeficiente de convección interno
h_i = Nu * k_i / D


# ============================================================
# 2. EXPRESIÓN DE T_out EN FUNCIÓN DE T_s
#    Solución de:
#    m_dot*cp*dT/dx = h_i*P*(T_s - T)
# ============================================================

beta = h_i * P * L / (m_dot * cp)

def T_out_from_Ts(T_s):
    """
    Temperatura de salida del fluido interno
    en función de la temperatura de pared T_s.
    """
    return T_s - (T_s - T_in) * np.exp(-beta)


# ============================================================
# 3. COEFICIENTE EXTERNO POR CONDENSACIÓN
#    Condensación laminar en película sobre cilindro horizontal
# ============================================================

def h_condensation(T_s):
    """
    Coeficiente promedio de condensación externo.
    T_s en °C.
    """
    deltaT = T_sat - T_s

    if deltaT <= 0:
        # Físicamente no tendría sentido para este problema
        return np.nan

    h_o = 0.725 * (
        rho_l * (rho_l - rho_v) * g * h_fg * k_l**3
        / (mu_l * D * deltaT)
    )**0.25

    return h_o


# ============================================================
# 4. ECUACIÓN NO LINEAL
#    q_condensacion = q_fluido
# ============================================================

def residual(T_s_array):
    """
    Residuo de la ecuación no lineal.
    fsolve trabaja con arrays, por eso recibimos T_s_array.
    """
    T_s = T_s_array[0]

    T_out = T_out_from_Ts(T_s)
    h_o = h_condensation(T_s)

    q_fluid = m_dot * cp * (T_out - T_in)
    q_cond = h_o * A * (T_sat - T_s)

    return [q_cond - q_fluid]


# ============================================================
# 5. SOLUCIÓN DE LA ECUACIÓN NO LINEAL
# ============================================================

# Suposición inicial razonable para T_s
T_s_guess = 90.0

T_s_solution = fsolve(residual, [T_s_guess])[0]

# Cálculos finales
T_out = T_out_from_Ts(T_s_solution)
h_o = h_condensation(T_s_solution)

q_fluid = m_dot * cp * (T_out - T_in)
q_cond = h_o * A * (T_sat - T_s_solution)


# ============================================================
# 6. RESULTADOS
# ============================================================

print("=" * 60)
print("RESULTADOS DEL PROBLEMA")
print("=" * 60)

print(f"Reynolds interno, Re = {Re:.2f}")
print(f"Nusselt interno, Nu = {Nu:.2f}")
print(f"Coeficiente interno, h_i = {h_i:.2f} W/(m^2·K)")
print()

print(f"Temperatura de pared, T_s = {T_s_solution:.4f} °C")
print(f"Temperatura de salida, T_out = {T_out:.4f} °C")
print(f"Coeficiente externo por condensación, h_o = {h_o:.2f} W/(m^2·K)")
print()

print(f"Calor ganado por el fluido, q_fluid = {q_fluid:.2f} W")
print(f"Calor cedido por condensación, q_cond = {q_cond:.2f} W")
print(f"Diferencia q_cond - q_fluid = {q_cond - q_fluid:.6f} W")
print("=" * 60)
```
